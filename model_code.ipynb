```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


df =pd.read_csv("/content/sample_data/novagen_dataset.csv")
df.head()
X=df.drop(["Target"], axis=1)
y=df["Target"]
import matplotlib.pyplot as plt

df['Target'].value_counts().plot(kind='bar')

plt.title('Target Variable Distribution')
plt.xlabel('Target')
plt.ylabel('Count')
plt.xticks([0,1], ['Unhealthy(0)','Healthy(1)'])
plt.show()
```

```python
import matplotlib.pyplot as plt

models = ['DT', 'LR', 'KNN', 'SVM', 'GBC', 'Voting', 'RF']
accuracy = [79.53, 81.36, 88.32, 93.35, 93.51, 92.67, 93.92]

bars = plt.bar(models, accuracy, color='skyblue')

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():.2f}%',
        ha='center'
    )

plt.ylabel('Accuracy (%)')
plt.title('Model Comparison')
plt.ylim(0, 100)
plt.show()
```

```python
# All your trained models and their names in a dictionary
trained_models = {
    'Decision Tree': DTclass,
    'Logistic Regression': LR,
    'KNN': KNN,
    'SVM': svc_model,
    'Gradient Boosting': gbc,
    'Voting Classifier': voting_clf,
    'Random Forest': RF
}

# Calculate recall and accuracy for each model automatically
model_names = []
recall_scores = []
accuracy_scores = []

for name, model in trained_models.items():
    y_pred = model.predict(X_test_scaled)
    model_names.append(name)
    recall_scores.append(recall_score(y_test, y_pred))
    accuracy_scores.append(accuracy_score(y_test, y_pred))

# Add newlines to long names so they fit under bars
display_names = [name.replace(' ', '\n') for name in model_names]

# ── Graph 1: Recall Comparison ────────────────────────────────────────────────
colors = ['#C44E52' if r == max(recall_scores) else '#AEC6CF' for r in recall_scores]
best_index = recall_scores.index(max(recall_scores))

fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(display_names, [r * 100 for r in recall_scores], color=colors)
ax.bar_label(bars, fmt='%.1f%%', fontsize=10, padding=3)
ax.set_ylim(70, 105)
ax.set_ylabel('Recall (%)')
ax.set_title('Which model is best at detecting Healthy patients?\n(Recall — Higher is Better)', fontsize=13)
ax.annotate(f'Best Model\n{model_names[best_index]}',
            xy=(best_index, max(recall_scores) * 100),
            xytext=(best_index - 2, 102),
            arrowprops=dict(arrowstyle='->', color='red'),
            fontsize=10, color='red')
plt.tight_layout()
plt.show()

# ── Graph 2: Accuracy Comparison ──────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 5))
bars = ax.bar(display_names, [a * 100 for a in accuracy_scores], color='#4C72B0')
ax.bar_label(bars, fmt='%.1f%%', fontsize=10, padding=3)
ax.set_ylim(70, 105)
ax.set_ylabel('Accuracy (%)')
ax.set_title('Overall Accuracy of Each Model', fontsize=13)
plt.tight_layout()
plt.show()

```
